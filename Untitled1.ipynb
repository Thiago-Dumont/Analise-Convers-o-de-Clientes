```python
import pandas as pd

# carregar o arquivo Excel
df = pd.read_excel('Dados.xlsx')

# visualizar os dados
df.head()
```

```python
df.columns
```

```python
# calcular conversões entre etapas

df['conv_visita_visualizacao'] = df['Visualizações'] / df['Visitas']
df['conv_visualizacao_sacola'] = df['Sacola'] / df['Visualizações']
df['conv_sacola_revisao'] = df['Revisão'] / df['Sacola']
df['conv_revisao_compra'] = df['Concluídos'] / df['Revisão']
df['conv_total_real'] = df['Concluídos'] / df['Visitas']

df[['conv_visita_visualizacao',
    'conv_visualizacao_sacola',
    'conv_sacola_revisao',
    'conv_revisao_compra',
    'conv_total_real']]
```

```python
df[['conv_visita_visualizacao',
    'conv_visualizacao_sacola',
    'conv_sacola_revisao',
    'conv_revisao_compra',
    'conv_total_real']].mean()
```

```python
import matplotlib.pyplot as plt

etapas = ['Visitas', 'Visualizações', 'Sacola', 'Revisão', 'Compras']
valores = [
    df['Visitas'].iloc[0],
    df['Visualizações'].iloc[0],
    df['Sacola'].iloc[0],
    df['Revisão'].iloc[0],
    df['Concluídos'].iloc[0]
]

plt.figure(figsize=(7,4))
plt.plot(etapas, valores, marker='o')
plt.title('Funil de Conversão da Loja')
plt.ylabel('Clientes')
plt.show()
```
